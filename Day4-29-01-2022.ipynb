While Loop

```python
import datetime
i=5
start_time = datetime.datetime.now()
print(start_time)
while i<10000:
    # if i%2 == 0 :
    #     i += 3
    #     print("The Value of i is ", i)
    #     continue
    print("The Value of i is ", i)
    i += 1
else :
    end_time = datetime.datetime.now()
    print(end_time)
    print("Total time taken for the loop : ", end_time - start_time)
```

Try Except Block

```python
try:
    i=10
    while i<100:
        print("The Value of i is ", i)
        i += 1
    else:
        raise ValueError("i is greater than 100")
except ValueError as error:
    print(error)
finally:
    print("All Done")
```
